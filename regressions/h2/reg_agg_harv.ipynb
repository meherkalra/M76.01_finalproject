```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
harv_agg = pd.read_csv("/content/harvard_aggregated_comment_class_sentiment.csv")
harv_centrality = pd.read_csv("/content/harvard_centrality_new.csv")
```

```python
c_dtypes = {'author':'str', 'btw_centrality':'float', 'deg_centrality':'float',
       'eig_centrality':'float', 'pagerank':'float', 'clustering_coeff':'float'}
harv_agg = pd.read_csv("/content/harvard_comment_sentiment_class.csv")
harv_agg['compound_sentiment'] = harv_agg['comment_sentiment_dict'].apply(lambda x: float(eval(x)['compound']))
harv_agg['positive_sentiment'] = harv_agg['comment_sentiment_dict'].apply(lambda x: float(eval(x)['pos']))
harv_agg['negative_sentiment'] = harv_agg['comment_sentiment_dict'].apply(lambda x: float(eval(x)['neg']))
harv_agg['comment'] = harv_agg['comment'].apply(lambda x: eval(x))
harv_agg['comment_author'] = harv_agg['comment'].apply(lambda x: x['author'])
harv_agg['comment_score'] = harv_agg['comment'].apply(lambda x: x['score'])
harv_agg.head(1)['comment'].apply(lambda x: print(x.keys()))
```

```python
merged_df = harv_centrality.merge(harv_agg[['comment_author', 'comment_score', 'positive_sentiment']], left_on='author', right_on='comment_author', how='left')
merged_df = merged_df.dropna()
```

```python
# X1 = cornell_agg['num_comments'].values
X2 = merged_df['comment_score'].values
X3 = merged_df['btw_centrality'].values
X4 = merged_df['deg_centrality'].values
X6 = merged_df['clustering_coeff'].values
y = merged_df['positive_sentiment'].values
preds = np.column_stack((X2,X3, X6))
labels = ["comment_score", "btw_centrality", "clustering_coeff"]
l = [X2, X3, X6]
```

```python
df = pd.DataFrame(preds, columns=labels)
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
#regression with uncorrelated coefficients 
p = np.column_stack((X2, X3))
#regression model
model = sm.OLS(y, sm.add_constant(p))
results = model.fit()
print(results.summary())
```
